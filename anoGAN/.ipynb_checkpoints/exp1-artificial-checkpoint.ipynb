### Purpose of Experiment

* 시계열데이터에 대해서 GAN을 학습하고
* 학습된 GAN이 다른 패턴의 변화를 탐지하는지 확인한다.

* 패턴의 변화(anomaly) 이란?
    - 시계열데이터는 trend, seanality, cycle, random noise의 성분으로 구성된다. 
    - 각 성분의 변화를 패턴의 변화로 정의하자

### Desing of  Experiment 

* anoGAN을 2가지 방식으로 학습

* 모델 학습에 사용할 데이터는 series1으로 시계열 데이터에서 전체 10000개의 샘플을 추출할 것임
    - Model1 : window-size만큼 1개의 샘플 추출. 샘플별로 overlap을 허용함.
    - Model2 : period_size(주기)만큼 1개의 샘플 추출. 샘플별로 overlap을 허용하지 않음


* 각 모델이 시계열 데이터의 trend, seasonality, noise의 변화를 잘 감지하는지 테스트


| |train data | ||chage detection|||
|---|---|---|---|---|
| |<b>sample length|<b>overlapping|<b><center> trend|<b><center>seasonality|<b><center>noise|
|<b>Model1 | period-size(1024) |<center>X| <center>0.57|<center> 1.0 | <center>0.97 |

### Generating artificial data

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

np.random.seed(100)
sample_len = 1024   
n_sample = 50000

def gen_series(sample_len, n_sample, trend_w, season_freq, noise_var):
    x = np.arange(0, sample_len*n_sample)
    trend = trend_w * x 
    seasonality = [np.sin(2 * np.pi * season_freq * (i/(sample_len*n_sample))) for i in np.arange(sample_len*n_sample)]
    noise = np.random.normal(0, noise_var, len(x))
    data = trend + seasonality + noise
    return data

# #### 기본 시계열 데이터 
series_standard = gen_series(sample_len, n_sample, 0, 50000, 0.3)
plt.figure(figsize=(15,3))
plt.plot(series_standard[:sample_len*10]) 
plt.show()

# #### anomaly 1 : trend shift - increase 0.5 within one period
series_trend = gen_series(sample_len, n_sample, 0.0001, 50000, 0.3)
plt.figure(figsize=(15,3))
plt.plot(series_trend[:sample_len*10]) 
plt.show()

#### anomaly 2 : frequency change - increase x1.5 
# series_seasonal = gen_series(sample_len, n_sample, 0, 50000*1.5, 0.3)
plt.figure(figsize=(15,3))
plt.plot(series_seasonal[:sample_len*10]) 
plt.show()

#### anomaly 3 : noise chage - increase to 0.5 from 0.3
series_noise = gen_series(sample_len, n_sample, 0, 50000, 0.5)
plt.figure(figsize=(15,3))
plt.plot(series_noise[:sample_len*10]) 
plt.show()

np.savetxt('series_standard.txt', series_standard, delimiter=',')
np.savetxt('series_trend.txt', series_trend, delimiter=',')
np.savetxt('series_seasonal.txt', series_seasonal, delimiter=',')
np.savetxt('series_noise.txt', series_noise, delimiter=',')
```

### Model2 : Overlap된 window별로 샘플 생성 후 학습

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split

np.random.seed(100)

def rolling_window(a, window):
    shape = a.shape[:-1] + (a.shape[-1] - window + 1, window)
    strides = a.strides + (a.strides[-1],)
    return np.lib.stride_tricks.as_strided(a, shape=shape, strides=strides)

series_standard = np.genfromtxt('series_standard.txt', delimiter=',')
series_standard = rolling_window(series_standard, 1024)
idx = np.random.randint(0, series_standard.shape[0],11000)
X_train = series_standard[idx[:10000], :]
X_test = series_standard[idx[10000:], :]
np.savetxt('X_train_window.txt', X_train, delimiter=',')

series_trend = np.genfromtxt('series_trend.txt', delimiter=',')
series_trend = rolling_window(series_trend, 1024)
idx = np.random.randint(0, series_trend.shape[0],1000)
X_test_1 = series_trend[idx, :]
X_test_1 = np.concatenate((X_test, X_test_1), axis=0)
y_test_1 = np.array([0] * 1000 + [1] * 1000).reshape(-1,1)  ##################### label for test
X_test_1 = np.concatenate((y_test_1, X_test_1), axis=1)
np.savetxt('X_test_1_window.txt', X_test_1, delimiter=',')

series_seasonal = np.genfromtxt('series_seasonal.txt', delimiter=',')
series_seasonal = rolling_window(series_seasonal, 1024)
idx = np.random.randint(0, series_seasonal.shape[0],1000)
X_test_2 = series_seasonal[idx, :]
X_test_2 = np.concatenate((X_test, X_test_2), axis=0)
y_test_2 = np.array([0] * 1000 + [1] * 1000).reshape(-1,1)
X_test_2 = np.concatenate((y_test_2, X_test_2), axis=1)
np.savetxt('X_test_2_window.txt', X_test_2, delimiter=',')

series_noise = np.genfromtxt('series_noise.txt', delimiter=',')
series_noise = rolling_window(series_noise, 1024)
idx = np.random.randint(0, series_noise.shape[0],1000)
X_test_3 = series_noise[idx, :]
X_test_3 = np.concatenate((X_test, X_test_3), axis=0)
y_test_3 = np.array([0] * 1000 + [1] * 1000).reshape(-1,1)
X_test_3 = np.concatenate((y_test_3, X_test_3), axis=1)
np.savetxt('X_test_3_window.txt', X_test_3, delimiter=',')

print(X_train.shape, X_test_1.shape, X_test_2.shape, X_test_3.shape)
```
